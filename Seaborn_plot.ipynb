```python
import seaborn as sns
```

```python
x= sns.load_dataset('tips')
```

```python
x
```

```python
sns.distplot(x['total_bill'])
```

```python
sns.jointplot(x='total_bill',y='tip',data=x,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=x,kind='resid')
```

```python
sns.pairplot(x,hue='sex')
```

```python
sns.rugplot(x['total_bill'])
```

```python
sns.barplot(x='sex',y='total_bill',data=x)
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=x,estimator=np.std)
```

```python
sns.countplot(x='sex',data=x)
#count the number in each group
```

```python
sns.boxplot(x='day',y='total_bill',data=x,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=x,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=x,hue='sex',jitter=True)
#jitter makes the distribution of our points better.
```

```python
sns.violinplot(x='day',y='total_bill',data=x)
sns.swarmplot(x='day',y='total_bill',data=x,color='k')
```

```python
i=sns.load_dataset('iris')
```

```python
import matplotlib.pyplot as plt

g= sns.PairGrid(i)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
p=sns.FacetGrid(data=x,col='time',row='smoker')
p.map(sns.distplot,'total_bill')
```
